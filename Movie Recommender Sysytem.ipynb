```python
import pandas as pd
import numpy as np
```

```python
# Load The Datasets 
```

```python
movies = pd.read_csv(r"tmdb_5000_movies.csv")
credits = pd.read_csv(r"tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# Merging Both Datasets
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies.head(1)
```

```python
# Feture Selection
```

```python
# genres
# id
# keywords
# overview
# title
# cast
# crew

dataset = movies[["id","title","overview","genres","keywords","cast","crew"]]
```

```python
dataset.info()
```

```python
# Filling NUll Values
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
# Removing duplicates Values
```

```python
dataset.duplicated().sum()
```

```python
dataset.head(1)
```

```python
# Converting Genres in Proper Format
```

```python
dataset["genres"].iloc[0]
```

```python
import ast
def convert(obj):
    li = []
    for i in ast.literal_eval(obj):
        li.append(i["name"])

    return li
```

```python
dataset["genres"] = dataset["genres"].apply(convert)
```

```python
dataset.head(1)
```

```python
dataset["keywords"] = dataset["keywords"].apply(convert)
```

```python
def convert2(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter += 1
        else:
            break
    return l
```

```python
dataset["cast"] = dataset["cast"].apply(convert2)
```

```python
def fetch_director(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
dataset["crew"] = dataset["crew"].apply(fetch_director)
```

```python
dataset.head(1)
```

```python
dataset["overview"] = dataset["overview"].apply(lambda x : x.split())
```

```python
# Removing The Spaces
```

```python
dataset["genres"] = dataset["genres"].apply(lambda x : [i.replace(" ","") for i in x])
dataset["keywords"] = dataset["keywords"].apply(lambda x : [i.replace(" ","") for i in x])
dataset["cast"] = dataset["cast"].apply(lambda x : [i.replace(" ","") for i in x])
dataset["crew"] = dataset["crew"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
dataset.head()
```

```python
dataset["tags"] = dataset["overview"] + dataset["genres"] + dataset["keywords"] + dataset["cast"] + dataset["crew"]
```

```python
new_dataset = dataset[["id","title","tags"]]
```

```python
new_dataset["tags"] = new_dataset["tags"].apply(lambda x:" ".join(x))
```

```python
new_dataset["tags"][0]
```

```python
# Stemming
```

```python
from nltk.stem.porter import PorterStemmer
pc = PorterStemmer()
```

```python
def Stem(text):
    li = []

    for i in text.split():
        li.append(pc.stem(i))

    return " ".join(li)
         
```

```python
new_dataset["tags"] = new_dataset["tags"].apply(Stem)
```

```python
# Vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words="english")
```

```python
vectors = cv.fit_transform(new_dataset["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Model
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommendetion(movie):
    movie_index = new_dataset[new_dataset["title"] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_dataset["title"][i[0]])
```

```python
recommendetion("Batman")
```

```python
import joblib
joblib.dump(new_dataset.to_dict(),"movies_dict.pkl")
joblib.dump(similarity,"similarity.pkl")
```
